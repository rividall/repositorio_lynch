```python
fig,ax = plt.subplots()

fruits = ['E', 'F', 'M', 'A', 'M', 'J', 'J', 'A', 'S', 'O', 'N', 'D']
counts = [40, 50, 10, 15,20,20,20,20,20,20,20,100]
bar_colors = ['#1abc9c', '#2ecc71', '#3498db', '#9b59b6','#34495e','#f1c40f','#e67e22','#e74c3c','#ecf0f1','#95a5a6','#16a085','#2980b9']

ax.bar(fruits, counts, width=0.9, color=bar_colors)

ax.set_ylabel('Cantidad de bombardeos')
ax.set_title('Bombardeos mensuales')

plt.show()
```
